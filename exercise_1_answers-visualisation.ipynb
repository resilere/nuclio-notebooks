```python
import plotly.express as px
from plotly.data import tips

df = tips()
df.head()
```

## 1. Create a histogram for the number of tip amounts

```python
px.histogram(df, 'tip', title='Histogram of Tip Amounts').show()
```

## 2. Create a new column, which will be tip/total_bill. Create a histogram of this column. 
## Is it the same shape as the histogram of tip amounts? If not, describe how it is different

```python
df['tip_over_bill'] = df['tip']/df['total_bill']
px.histogram(df, x='tip_over_bill', title='Histogram of tip/bill').show()
# The histogram is more bell-shaped. The histogram of tips has a longer tail. 
# This shows that the tip amount scales with the bill in most instances, which is what we would expect
```

## 3. Create a visualisation to show the number of tips per day. What is the most common day for tips? 
##   (hint: You will need to use groupby and count the number of rows each day)

```python
dft = df.groupby('day').tip.count().reset_index()

category_orders = {
    'day': ['Thur', 'Fri', 'Sat', 'Sun'],
    'sex': ['Male', 'Female']
}

px.bar(dft, x='day', y='tip', text='tip', title='Number of Tips Per Day', 
       category_orders=category_orders).show()
```

## 4. Create another visualisation showing the average tip amount per day. 
##   You should use a different plot type to the one used to answer the question above. 
##   Which day has the highest average tip amount?

```python
dft = df.groupby('day').tip.mean().reset_index()
dft.tip = dft.tip.round(2)
dft = dft.sort_values('tip', ascending=False)

fig = px.scatter(dft, x='day', y='tip', text='tip', title='Average Tips Per Day')
fig.update_traces(textposition='middle left', marker_size=20)
fig.show()

fig2 = px.pie(dft, values='tip', names='day', title='Average Tips Per Day')
fig2.show()
```

## 6. Create a visualisation to show which gender is most likely to tip.

```python
dfg = df.groupby('sex').tip.count().reset_index()
px.bar(dfg, x='sex', y='tip', text='tip', title='Most likely to tip, split by sex')
```
