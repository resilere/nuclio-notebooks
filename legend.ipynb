```python
import plotly.express as px

df = px.data.gapminder().query("year==2007")
fig = px.scatter(df, x="gdpPercap", y="lifeExp", color="continent",
    size="pop", size_max=45, log_x=True)
fig.show()
```

The legend appears automatically. We can update the legend layout using the legend attribute of layout.

```python
fig.update_layout(
    legend=dict(
      yanchor="top",
      xanchor="left",
      y=0.99,
      x=0.01
  )
)

fig.show()
```

We can also turn the legend off

```python
fig.update_layout(showlegend=False)
fig.show()
```

We can also add/modify the legend title

```python
fig.update_layout(showlegend=True)
fig.update_layout(legend_title_text = 'Continents of the World')
fig.show()
```
