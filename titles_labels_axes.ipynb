```python
import plotly.graph_objects as go
```

1. We can name each set of data we add using the name attribute of any GO plot object



```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=[0, 1, 2, 3, 4, 5, 6, 7, 8],
    y=[0, 1, 2, 3, 4, 5, 6, 7, 8],
    name="Line 1" # Name of the line, not the axis
))
fig.show()
```

```python
fig.add_trace(go.Scatter(
    x=[0, 1, 2, 3, 4, 5, 6, 7, 8],
    y=[2, 3, 4, 3, 2, 3, 4, 5, 7],
    name="Line 2"
))
fig.show()
```

```python
print(fig)
```

2. We can add titles using update_layout()

```python
fig.update_layout(title_text = 'This is the title')

# fig.update(layout_title_text = 'This is also a title') # Example of magic notation
fig['layout']['title']['text'] = 'This also works'

fig.show()
```

3. We add axes using either update, update_layout of update_xaxes/update_yaxes

```python
fig.update_xaxes(title_text= 'Time')
#fig['layout']['xaxes']['title']['text'] = 'Time'
fig.update_yaxes(title_text = 'Value')
fig.show()
```

4. We can change the fonts and colours of any of these by editing their font objects

```python
fig.update_layout(title_font_size = 20)
fig.update_layout(title_font_color = 'red')
fig.update_layout(xaxis_title_font_color = 'blue')
fig.update_layout(yaxis_title_font_size = 5)
fig.show()
```

```python
print(fig)
```

5. We can change the size and colour of all labels that have not been modified by updating font. 

```python
fig.update_layout(font_size=20)
fig.update_layout(font_color='green')
```

Remember there are usually lots of ways to do the same thing. The cleanest way to modify the plot is probably to update everything inside of an update_layout call, but its really up to you.

```python
fig.update_layout( 
    title="Plot Title",
    xaxis_title="X Axis Title",
    yaxis_title="Y Axis Title",
    # You can also play with the fonts
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"
    )
)

fig.show()
```
